```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt


file_path = '/mnt/c/preprocessing/brain_masking_created/HF3_G1_ok/T2.nii'
image = nib.load(file_path)


image_data = image.get_fdata()
image_dimensions = image_data.shape
voxel_size = image.header.get_zooms()
affine_matrix = image.affine
data_type = image_data.dtype
intensity_range = (np.min(image_data), np.max(image_data))

image_properties = {
    "Image Dimensions": image_dimensions,
    "Voxel Size": voxel_size,
    "Affine Matrix": affine_matrix,
    "Data Type": data_type,
    "Intensity Range": intensity_range
}

image_properties
```

```python
def plot_histogram(image_data, title=""):
    plt.hist(image_data.flatten(), bins=256, color='blue', alpha=0.7)
    plt.title(title)
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.show()

# Load the NIfTI file
file_path = '/mnt/c/preprocessing/brain_masking_created/HF3_G1_ok/T2.nii'  # Replace with the path to your NIfTI file
image = nib.load(file_path)
image_data = image.get_fdata()

# Plot histogram
plot_histogram(image_data, title="Histogram of Image")
```
