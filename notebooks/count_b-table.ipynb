```python
def generate_b_table_v2(input_file_path, output_file_path):
    with open(input_file_path, 'r') as file:
        lines = file.readlines()

    dw_dir = []
    dw_bval = None
    for index, line in enumerate(lines):
        if "##$PVM_DwDir=" in line:
            for _ in range(int(line.split()[1].strip(','))):
                index += 1
                dw_dir.extend(lines[index].strip().split())
        if "##$PVM_DwBvalEach=" in line:
            dw_bval = int(lines[index + 1].strip())

    formatted_output = ["0   0   0   0"] * 9
    for i in range(0, len(dw_dir), 3):
        formatted_output.append(f"{dw_bval}   {'   '.join(dw_dir[i:i+3])}")

    with open(output_file_path, 'w') as out_file:
        out_file.write('\n'.join(formatted_output))


generate_b_table_v2("../methods/method_hf3_g4", "../methods/b-table.txt")
```
