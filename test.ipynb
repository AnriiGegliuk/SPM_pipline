```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = nib.load('/mnt/c/preprocessing/brain_masking_created/HF3_S2_ok/3rd/brain_mask_filled.nii')
data = img.get_fdata()
```

```python
np.min(data), np.max(data)
```

```python
data_flat = data.flatten()

plt.figure(figsize=(10,6))
plt.hist(data_flat, bins=50, color='c', edgecolor='k', alpha=0.7)
plt.title('Histogram of Pixel Intensities in the Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
def display_slices(data):
    axial_middle = data.shape[2] // 2
    coronal_middle = data.shape[1] // 2
    sagittal_middle = data.shape[0] // 2
    
    axial_slice = data[:, :, axial_middle]
    coronal_slice = data[:, coronal_middle, :]
    sagittal_slice = data[sagittal_middle, :, :]
    
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(axial_slice.T, cmap='gray', origin='lower')
    axes[0].set_title('Axial')
    axes[0].axis('off')
    
    axes[1].imshow(coronal_slice.T, cmap='gray', origin='lower')
    axes[1].set_title('Coronal')
    axes[1].axis('off')
    
    axes[2].imshow(sagittal_slice.T, cmap='gray', origin='lower')
    axes[2].set_title('Sagittal')
    axes[2].axis('off')
    
    plt.show()
```

```python
display_slices(data)
```
